```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
btc_data = pd.read_csv('..\BTCUSDT_1d_all_data.csv')
print(btc_data.head()) 
```

```python
plt.plot(btc_data['Open Time'], btc_data['Close'], linestyle='-')
plt.xlabel('Open Time')
plt.ylabel('Close Price')
plt.title('BTCUSDT Close Price Over Time')
plt.show()
```

```python
nan_rows = btc_data[btc_data.isna().any(axis=1)]
print(f"Number of rows with NaN values: {len(nan_rows)}")
```
